```
import numpy as np
n=int (input('number_no_nodes '))
arr=np.arange((n-2)*(n-2)).reshape(n-2,n-2)*0
arr=arr.astype(float)
arr
```

```
alpha=0.020875
# left hand side
for i in range(n-2):
 arr[i][i]=1+(2*alpha)
for i in range(3):
 arr[i][i+1]=-alpha
for i in range(1,4):
  arr[i][i-1]=-alpha
arr
```

```
T_first_node_0=100 #Boundary conditons
t_last_node_n=50
```

```
# Right hand side
d=d=np.arange(n-2).reshape(n-2,1)*0
d=d.astype(float)
d[0][0]=alpha*(T_first_node_0)
d[n-3][0]=alpha*(t_last_node_n)
d
```

```
solution=np.arange(n-2).reshape(n-2,1)*0  #Intially
solution=solution.astype(float)
solution
```

```
time=int (input('time '))
for i in range(time):
 solution=np.linalg.solve(arr,(d+solution))
 print('At t= ',i)
 print(solution)
```
